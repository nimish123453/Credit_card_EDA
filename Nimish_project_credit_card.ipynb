```python
#through iqr method simply remove the outliers by removing the data points which are above upper limit and below lower limit from the total data
for i in df.columns:
  q1=df[i].quantile(0.25)
  q3=df[i].quantile(0.75)
  iqr=q3-q1
  LL=q1-1.5*iqr
  UL=q3+1.5*iqr
  print('IQR of',i,'=',iqr)
  print('LL of',i,'=',LL)
  print('UL of',i,'=',UL)
  print('Number of outlair in',i,'=',(df.shape[0]-df[(df[i]<UL) & (df[i]>LL)].shape[0]))     #df.shape[0] #no of rows
  print(' ')
```
